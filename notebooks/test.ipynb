```python
import os
from dotenv import load_dotenv
import psycopg2
from src.infrastructure.database.connection_service import PostgreSQLDatabaseConnectionService

# Load environment variables from .env file
load_dotenv()

# Get the database path from environment variables
database_path = os.getenv("DATABASE_PATH")

if not database_path:
    raise ValueError("DATABASE_PATH not found in environment variables. Please set it in your .env file.")

# Create an instance of the database connection service
db_service = PostgreSQLDatabaseConnectionService(db_path=database_path)

try:
    # Get a raw database connection
    connection = db_service.get_raw_connection()

    # Create a cursor
    cursor = connection.cursor()

    # Define and execute a SQL query to list tables
    query = "SELECT table_name FROM information_schema.tables WHERE table_schema = 'public';"
    cursor.execute(query)

    # Fetch all the results
    tables = cursor.fetchall()
    print("Tables in the database:")
    for table in tables:
        print(table[0])

    # Close the cursor
    cursor.close()

except (Exception, psycopg2.Error) as error:
    print("Error while connecting to PostgreSQL", error)

finally:
    # Close the database connection
    if 'connection' in locals() and connection is not None:
        db_service.close_connection()
        print("PostgreSQL connection is closed")
```
